```python
import pandas as pd 
import re
```

```python
data = pd.read_csv('../api/data/sample_client_data.csv')
data.drop(columns="Unnamed: 0", inplace=True)
data.head()
```

In order to ease the retrieval and display of information, we'll reverse the encoding of the categorical variables

```python
encoded_columns = data.loc[:, 'CODE_GENDER_F':'WALLSMATERIAL_MODE_Wooden'].columns
encoded_feature_names = set(re.findall(r'(?<![a-z])[A-Z_]{2,}_', ' '.join(encoded_columns)))
encoded_feature_names
```

```python
for feature in encoded_feature_names:
    data[feature[:-1]] = data.filter(regex=f'^{feature}', axis=1).idxmax(axis=1).apply(lambda x: x[len(feature):])
```

```python
data.filter(regex=f'^NAME_FAMILY')
```

```python
data.drop(columns=encoded_columns, inplace=True)
```
