```python
import pandas as pd
import re
```

```python
historical_data = pd.read_csv("../api/data/sample_historical_data.csv")
historical_data.drop(columns="Unnamed: 0", inplace=True)


data = pd.read_csv("../api/data/sample_client_data.csv")
data.drop(columns="Unnamed: 0", inplace=True)
data.head()
```

```python
data["DAYS_BIRTH"] = abs(data["DAYS_BIRTH"])
data["DAYS_EMPLOYED"] = abs(data["DAYS_EMPLOYED"])
historical_data["DAYS_EMPLOYED"] = abs(historical_data["DAYS_EMPLOYED"])
historical_data["DAYS_EMPLOYED_PERCENT"] = abs(historical_data["DAYS_EMPLOYED_PERCENT"])

data.to_csv("../api/data/sample_client_data.csv")
historical_data.to_csv("../api/data/sample_historical_data.csv")
```

In order to ease the retrieval and display of information, we'll reverse the encoding of the categorical variables

```python
encoded_columns = data.loc[:,
                           'CODE_GENDER_F':'WALLSMATERIAL_MODE_Wooden'].columns
encoded_feature_names = set(re.findall(
    r'(?<![a-z])[A-Z_]{2,}_', ' '.join(encoded_columns)))
encoded_feature_names
```

```python
for feature in encoded_feature_names:
    data[feature[:-1]] = data.filter(regex=f'^{feature}', axis=1).idxmax(
        axis=1).apply(lambda x: x[len(feature):])
```

```python
data.filter(regex=f"^NAME_FAMILY")
```

```python
data.drop(columns=encoded_columns, inplace=True)
```

```python
for col in data.columns:
    if data[col].dtype == float:
        data[col] = data[col].astype(int)
```
